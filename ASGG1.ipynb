```python
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split as split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import  RobustScaler

df = pd.read_csv('Inequality in Education.csv')

df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Country','UNDP Developing Regions' ])
```

```python

# Replace NaN values with the mean for each column
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):  # Check if the column is numeric
        df[col] = df[col].fillna(df[col].mean())
# Print the first 10 rows
df.head(10)
```

```python
df['Human Development Groups'] = df['Human Development Groups'].fillna('NaN')

# Create a mapping from human development group to numerical values.
human_development_mapping = {
    "NaN": 0,
    "Low": 1,
    "Medium": 2,
    "High": 3,
    "Very High": 4
}

# Apply the mapping to the "Human Development Groups" column.
df["Human Development Groups Numerical"] = df["Human Development Groups"].map(human_development_mapping)
df = df.drop(columns=["Human Development Groups"]) #drop column HDG
df.head()
```

```python
# prompt: Using dataframe df: ISO 3 convert to numerical order

# Create a mapping of ISO3 codes to their numerical order
iso3_to_numerical = {iso3: i for i, iso3 in enumerate(df['ISO3'].unique())}

# Create a new column with the numerical order
df['ISO3_numerical'] = df['ISO3'].map(iso3_to_numerical)
df = df.drop(columns=["ISO3"]) #drop column HDG
# Display the dataframe to show the new column
df
```

```python
X = df.drop(columns=["Human Development Groups Numerical", ])
y = df['Human Development Groups Numerical']

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
```

```python
scaler = RobustScaler()
X_train_RRscaled = scaler.fit_transform(X_train)
X_test_RRscaled = scaler.transform(X_test)

print("Robust Scalar")
print(' ')
for x in range(0, 100, 10):
  lnr = Ridge(alpha=x).fit(X_train_RRscaled, y_train)
  print(f"Ridge alpha = {x} : {lnr.score(X_test_RRscaled, y_test):.3f}")
```
